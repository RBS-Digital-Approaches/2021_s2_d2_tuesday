# Visualizing a network with Kumu
[Kumu](https://kumu.io) is an online service that makes it pretty easy to create attractive, interactive network visualizations. While there is a paid tier available, you can do a lot with a cost-free account.

Rather than providing a detailed walk-through of using Kumu (which would basically just duplicate [their documentation](https://docs.kumu.io/)), I'll just touch on a couple of things that you should look out for and try, with links to the appropriate spots in the documentation.

## A - Before you begin
You'll need to [set up a free account](https://kumu.io/join) to use Kumu. (I've had an account for several years now, and can report that they don't spam my inbox.)

## 1 - Creating a project and choosing a layout
When you [start a new project](https://docs.kumu.io/getting-started/first-steps.html#create-your-first-project), you're immediately faced with a decision about what *kind* of graph to create. Because the networks that we're visualizing contain so many nodes, I would strongly recommend the "SNA" (Social Network Analysis) template.

## 2 - Getting your data into Kumu
You can import data about a network into Kumu by uploading an Excel or .csv file, or by pointing Kumu to a Google Sheet. 

If you've gotten here by following the Google Sheets path, then simply pointing Kumu to your Google Sheet is probably the easiest approach. If you've gotten here after preparing your node and edge lists in Python code, you'll need to upload the .csv files you created (or import those .csv files into a Google Sheet and then give Kumu that link)

There's a detailed guide to [importing data](https://docs.kumu.io/guides/import.html) in Kumu's documentation.

## 3 - Performing calculations
Once your data are imported, Kumu will create a visualization that you could set to customizing right away. I think it's worth your while, though, to first use Kumu's "Metrics" tool to [perform some analyses](https://docs.kumu.io/guides/metrics.html) of your network.

On the one hand, stopping to look at these metrics is a good reminder that the visualization isn't the network, itself, but rather a *view* of the network. It's the information about connections among the nodes in our network that's really valuable, and it's worth exploring and understanding that information a bit up front so that we can decide what kinds of visualization best serve our data.

At a practical level, having Kumu perform those analyses greatly expands what you can do in styling your visualization (by, for example, changing the size of nodes based on their "degree"â€”the number of connections they have to other nodes). 

## 4 - Styling the network
Once you get the hang of Kumu's interface, you'll find that there are lots of options for styling the visualization of your network available in the Basic editor (even more is possible if you decide to dive into entering some CSS-like code in the Advanced editor). 

Keeping track of where things are in Kumu's interface can be a little challenging, at first, but you can always get to the Basic editor by clicking on the Settings icon at the upper right of the visualization:

Your visualization will slide over to reveal the Basic editor. To dismiss the editor and allow your visualization to fill the whole window, just click on the Settings icon again.

Have a look at the [documentation for styling the visualization](https://docs.kumu.io/guides/data-driven-decorations.html#decorate-in-the-basic-editor) and experiment with making your visualization highlight the kinds of information you want it foreground for users.

## 5 - Exploring the network
Kumu allows you to explore and examine different parts of your network by filtering which nodes you see; by focusing on specific nodes or groups of nodes; by showcasing portions of the network; and by controlling how elements are clustered with one another. 

The documentation on [how to interact with the graph](https://docs.kumu.io/getting-started/first-steps.html#refine-your-view) to do these things is perhaps a little bit sparse, but should be enough to get you started.

One thing you might try would be first to size your nodes by degree (one of the network algorithms that you can calculate) and then click on a large node to bring it into focus. From there, you can expand the focus one layer at a time to see what else is connected to the node you've selected. (You can reset the focus by clicking on the Focus crosshair icon again.)

You can also select multiple nodes to focus on by holding down the shift key as you click on nodes. You might try clicking on two (or more) book nodes to see the readers they have in common (which will also bring into view the other books that those readers borrowed). You might try clicking on two (or more) borrower nodes to see the books they have in common (which will also bring into view other readers who borrowed those books).


## Next steps
Kumu makes it easy to create very nice visualizations, and it's definitely not a mere toy: you can do real analysis with it that you can share with others or embed on a web site.

The downside to using Kumu is that we're reliant on somebody else's service: if Kumu goes away, decides to start charging for what's currently a free service, or simply makes changes to their code, our visualization might become unavailable, or might not behave the way we expect. 

If you need to have more control over your network visualizations, one tool that a lot of people use is a desktop application called [Gephi](https://gephi.org). Gephi's web site describes the app as "Like Photoshop for graphs," which is pretty accurate. Often, users will use Gephi to prepare and explore a network visualization, then export a particular view of it as a static image for use in a publication. It's [possible](https://blog.miz.space/tutorial/2020/01/05/gephi-tutorial-sigma-js-plugin-publishing-interactive-graph-online/) to get a version of a graph made with Gephi online in an interactive way, but that's an area where there seems to be work left to do. 
  
 Most interactive network graph visualizations that you see online are rendered using one of a variety of [JavaScript libraries](https://elise-deux.medium.com/the-list-of-graph-visualization-libraries-7a7b89aab6a6). If you wanted to incorporate network visualization into a public-facing web project, it would be worth looking into those libraries.

The fundamental part of the work of putting together a network to visualize, though, is assembling the network dataâ€”the list of nodes and edges. That means the techniques you tried out in the earlier workbooks are worth investigating and interrogating: what is the kind of information you want to capture about your nodes? How do you want to think about the edges?
