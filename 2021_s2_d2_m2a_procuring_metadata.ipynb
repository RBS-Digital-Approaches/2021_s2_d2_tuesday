# Doing more with bibliographic metadata
There's a lot of bibliographic information packed into library catalogue records. But because those records were primarily designed to be presented one at a time to users of libraries online public access catalogs (OPACs), that information isn't quite as actionable as it could be. Today's units explore repurposing catalogue information in order to get a different view of bibliographical metadata from the one we get when clicking through records in a web browser.

For these metadata exercises, we want you to work with catalogue records for items that actually interest you, if possible. 
* For the first exercise, we'll want records with a range of publication places, so consider whether there's a catalogue search you could perform that would return books that don't all come from the same place. 
* For the second exercise, we want to see if we can work out networks of relationships reflected in imprints, so check to see whether the catalogue records you're looking at provide reasonably full imprint statements. 

If no topic springs to mind, if the topic you're really interested in doesn't lend itself to those criteria (because, say, you're interested in books that were all printed in one place), or if it turns out that the records you'd ordinarily want to use would require further processing to make them work, there's a sample collection of MARC records drawn from the English Short Title Catalogue you can use.



## Getting your metadata
Before we can repurpose bibliographic metadata from catalogue records, we have to have some bibliographic metadata from catalogue records to work with. In the interests of time, we'll stick to one or the other of two methods:

1. Saving catalogue records to Zotero
2. Bulk exporting records from a catalogue

If you are getting records for use with OpenRefine, you'll want to make sure that you are getting records in a form that Zotero can accept (don't worry, Zotero can handle lots of common formats), because we'll be using Zotero to move our catalogue records into .csv format.

If you are getting records to process with Python, you'll want either MARC or MARCXML.



### Saving catalogue records to Zotero
[Zotero](https://zotero.org) is a cross-platform, cost-free, open-source reference and citation manager and PDF organizer that was developed at the Roy Rosenzweig Center for History and New Media at George Mason University, and is now maintained by the nonprofit Corporation for Digital Scholarship.

Zotero fulfills the same function as similar software like EndNote or Mendeley, and is handy for saving citation information for books, archival sources, journal articles, web sites, and more. For our purposes, the appeal of Zotero lies in the fact that it's fully cross-platform (working on Windows, Mac, and Linux) and that it's open-source, so that it can be extended or customized. 

If you haven't installed Zotero already, you can [download it here](https://www.zotero.org/download/). Be sure to get both the Zoter application and the connector for your web browser (Mac users: Zotero notes that the Safari connector is not really ready for prime time at the moment. Stick with the connector for Chrome, which also works with other Chromium-based browsers, like Brave.)

We'll be using Zotero for its ability to bring in records from pretty much any library catalogue and export them in .csv format.

#### Zotero site translators
Zotero seems to handle pretty much anything you can throw at it, but if you are interested in getting records from the English Short Title Catalogue, the American Antiquarian Society catalogue, or from the Short Title Catalogue-Netherlands, download and install the site-specific translators you'll find at [https://github.com/benjaminpauley/Zotero-Translators](https://github.com/benjaminpauley/Zotero-Translators). Those translators will pull in some information (like full pagination statements) that Zotero wouldn't get by default.

#### Using Zotero
Zotero's browser connectors allow you to save information to Zotero as you go about your research. When the Zotero connector is installed and active in your browser, you'll see a new icon that will allow you to add items to Zotero (the icon will change depending on what kind of resource Zotero thinks you're looking at. 

You can read up on [adding items to Zotero](https://www.zotero.org/support/adding_items_to_zotero), then visit any library catalogue (or library catalogue-adjacent) site and search/browse as normal, adding results as you go.

Note that Zotero browser connector will add results to the [collection](https://www.zotero.org/support/collections_and_tags#collections) that's currently active in Zotero. It might be a good idea to create a new collection for an exercise before you begin browsing to keep your results organized.

### Bulk exporting/downloading from a catalogue
In addition to importing records directly from the browser, Zotero can bulk import records from [several different file types](https://www.zotero.org/support/kb/importing_standardized_formats).

Many library catalogues provide some kind of bulk export and download facility that will produce a file that Zotero can read, but it can take some looking around and experimenting to figure out what will work best. 

Many catalogues will allow you to export files designed for use with reference managers like Zotero (though they may not mention Zotero by name). These formats include RIS, BibTex, and EndNote, all of which Zotero can handle. Different catalogues may implement this functionality differently, so just be on the lookout for anything that seems to allow you to email, download, or save records. 



Different libraries' catalogues may handle things differently, even if they're running the same software.

#### Importing downloaded metadata into Zotero
Whatever format your metadata comes in, the process for getting it into Zotero is the same: select `Import` from the `File` menu...
import_to_Zotero.png


Click "Continue"...

import_to_Zotero_2.jpg

Navigate to your downloaded file (in your Downloads folder, or wherever else you tend to save such things) and click "OK." 

import_to_Zotero_3.jpg

By default, Zotero will put these imported items into a new collection. If you don't want it to do that, uncheck the radio button. Click "Continue."

import_to_Zoter_4.png

After that, there's one more dialog box (that I didn't capture), but all you need to do is click "Done" and you'll see your new items in Zotero.
