It's worth dwelling a little on Wikipedia's captioning of that image as a "*visualization* of a social network graph" (emphasis added). These kinds of images are very impressive (and, especially, when they're made interactive, can become entirely engrossing). But the images, themselves, are not the "network graph;" rather, they provide a visual *representation* of the graph, which consists of a collection of identifiable entities and the connections among them. (Which presumes that there *are* connections among them.)


## Nodes and edges
In network parlance, the entities in a network are generally referred to as "nodes," and the connections between them are called "edges." 

Network graphs can be either "directed" or "undirected." In a directed graph, there is a definite directionality in the connection—there is a "source" and a "target," and it matters which node is the source and which is the target. (If we wanted to represent a network of correspondence, for instance, we would care who was the sender of the letter and who was the recipient.) Undirected graphs don't really presume that kind of directionality: the edge is simply a statement that two nodes are connected, and that connection is presumably symmetrical (if two people are friends, each is a friend to the other).

As is often the case with digital work, perhaps the biggest part of the job is preparing the data that we want to work with. In order to construct a newtwork graph, we need to prepare:

* A list of all of the distinct entities that form nodes in the network
* A list of all of the connections between those nodes—the edges—with one edge per line.

Moreover, both raise the same kinds of questions about how we think about the connections between nodes in a network, and both involve considerations about possible tradeoffs between completeness and legibility when it comes to visualizing a network graph.
